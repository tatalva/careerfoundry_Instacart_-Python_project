## Task 3 -  Compare customer behavior in different geographic areas

```python
## Check the states
df_order_customer_prd['STATE'].value_counts()
```

```python
# create US region grouping the STATE as per Wikepedia
region=[]
for value in df_order_customer_prd['STATE']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
# create region column
df_order_customer_prd['region'] = region
```

```python
region
```

```python
df_order_customer_prd.shape
```

```python
## Check region
df_order_customer_prd['region'].value_counts()
```

```python
## Crosstab the region with spending flag
spending_region  = pd.crosstab(df_order_customer_prd['region'], df_order_customer_prd['spending_flag'], dropna = False)
```

## Task 4 - Exclude customers with low activity

```python
# Create a column for high activy customer - more than or equal to 5 orders max
df_order_customer_prd.loc[df_order_customer_prd['max_order'] >= 5, 'customer_activity'] = 'High-activity'
```

```python
# Create a column for low activy customer - lower than 5 orders max
df_order_customer_prd.loc[df_order_customer_prd['max_order'] < 5, 'customer_activity'] = 'Low-activity'
```

```python
## Check customer_activity
df_order_customer_prd['customer_activity'].value_counts()
```

#We will create a subset with 30964564 rows

```python
# Create a subset for only the High-activity customers
df_order_customer_prd_high = df_order_customer_prd[df_order_customer_prd['customer_activity'] == 'High-activity']
```

```python
#export fjle
df_order_customer_prd_high.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'full_file_v3.pkl'))
```
