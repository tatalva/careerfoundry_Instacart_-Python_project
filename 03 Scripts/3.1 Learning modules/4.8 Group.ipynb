## Create subset for fisrt million rows

```python
#create substet of 100000 rows to avoid memory issues
df = ords_prods_merg[:1000000]
```

## 1st step Slipt data into groups

```python
#group by product name. This function create an object
df.groupby('product_name')
```

## 2nd step Apply a function to each group separately.

```python
## Create grop by department_id and agreegated by order number
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
## we can also do this withoutht aggregation 
df.groupby('department_id')['order_number'].mean()
```

```python
## perfom mult aggregation
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## 3rd step create a flag - Create a flag for loyalty customers that by the same product

```python
ords_prods_merg['max_order'] = ords_prods_merg.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merg.head()
```

```python
ords_prods_merg.head(100)
```

## Now flag the column using loc

```python
ords_prods_merg.loc[ords_prods_merg['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merg.loc[(ords_prods_merg['max_order'] <= 40) & (ords_prods_merg['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merg.loc[ords_prods_merg['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merg['loyalty_flag'].value_counts(dropna = False)
```
